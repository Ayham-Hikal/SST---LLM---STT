# **2.) STT, LLM, & TTS**

```python
def GenAI(audio_file,output = "GenAI.mp3"):

  Record = sr.Recognizer()
  with sr.AudioFile(audio_file) as source:
    audio = Record.record(source)
    question = Record.recognize_google(audio)
  print(f"So your question is: {question}")

  genai.configure(api_key = ("YOUR API KEY GOES HERE")) # DO NOT FORGET TO ADD AN API KEY
  model = genai.GenerativeModel("gemini-2.5-flash")
  response = model.generate_content(question)
  answer  = response.text
  print(f"The answer is: {answer}")

  tts = gTTS(text=answer, lang='en')
  tts.save(output)
  print(f"\nFile is saved to {output}")
  return question, answer, output
```
